```python
for i in range(1,288):
    for j in range(1,432):
        im[i,j] = i*j + 2*j
        #im[i,j,1] = i*j + 2*j
        #im[i,j,2] = i*j + 2*j
plt.figure(figsize = (10,10))
plt.imshow(im, cmap = plt.get_cmap(name = 'gray'))
```

```python
Q = 2
for i in range(1,288):
    for j in range(1,432):
        im[i,j] = np.abs( np.cos(i/Q) + 2*np.sin(j/Q))
        #im[i,j,1] = i*j + 2*j
        #im[i,j,2] = i*j + 2*j
plt.figure(figsize = (10,10))
plt.imshow(im, cmap = plt.get_cmap(name = 'gray'))
```

```python
#f(x, y) = |3(x/Q) p3y/Q|;
Q = 100
for i in range(1,288):
    for j in range(1,432):
        im[i,j] = np.abs( 3 * (i/Q) - np.power((j/Q),0.3333333) )
        #im[i,j,1] = i*j + 2*j
        #im[i,j,2] = i*j + 2*j
plt.figure(figsize = (10,10))
plt.imshow(im, cmap = plt.get_cmap(name = 'gray'))
```
