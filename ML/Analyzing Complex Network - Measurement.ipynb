```python
##Code taken from NetworkX Documentation
##Compiled by jecs89

#Import lines
%pylab inline
import networkx as nx

from matplotlib import pyplot as plt
```

```python
prefix = './exp 1/'
neg = 'net_neg_'
pos = 'net_pos_'
pofix = '.net'

for i in range(1,6):
    G1=nx.read_pajek(prefix+neg+str(i)+pofix)
    G=nx.Graph(G1)
#     print(G.number_of_nodes())
#     print(G.number_of_edges())

    # print(nx.average_neighbor_degree(G))
    var = nx.average_neighbor_degree(G, weight='weight')
    mylist = []
    myval = []
    # print(var)
    for idx in var:
        mylist.append(int(idx))
        myval.append(var[idx])
    # print(mylist)
    # print(myval)
    
    ax = plt.subplot(3,2,i)
    ax.set_ylim((0, 8))
    plt.plot(mylist,myval)
```
