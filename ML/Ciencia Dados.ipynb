```python
#count, division = np.histogram(mydf['BedroomAbvGr'])
#plt.hist(count)

#count, division = np.histogram(mydf['Street'])
#plt.hist(count)

counter = 1
plt.figure(figsize = (7,7))
for column in mydf:
    plt.subplot(4,3,int(counter))
    plt.title(column)
    count, division = np.histogram(mydf[column])
    plt.subplots_adjust(hspace=0.5)
    plt.hist(count)
    counter += 1
plt.show()
plt.savefig('hist.png')
```

```python
counter = 1
plt.figure(figsize = (7,7))
for column in mydf:
    plt.subplot(4,3,int(counter))
    plt.title(column)
    plt.subplots_adjust(hspace=0.5)
    mydf[column].value_counts().plot.bar()
    counter += 1
plt.show()
```

```python
import math
k    = round(math.sqrt(mydf.shape[0]))
cols = ['amplitud']
amplitud = pd.DataFrame(index = mydf.columns.values, columns =cols)
amplitud = amplitud.fillna(0)
for column in mydf:
    val_max = round(mydf[column].max())
    val_min = round(mydf[column].min())
    temp    = (val_max - val_min) / k
    amplitud.loc[column] = temp
print('xxx')
print(amplitud)
```
